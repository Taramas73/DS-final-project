```python
import json
import numpy as np
import rasterio
from rasterio.features import rasterize
from shapely.geometry import shape
import matplotlib.pyplot as plt

def load_geojson(file_path):
    with open(file_path, 'r') as f:
        geojson_data = json.load(f)
    return geojson_data

def create_mask(geojson_data, width, height, bounds):
    geometries = [(shape(feature["geometry"]), 1) for feature in geojson_data["features"]]
    
    transform = rasterio.transform.from_bounds(*bounds, width, height)
    mask = rasterize(
        geometries,
        out_shape=(height, width),
        transform=transform,
        fill=0,
        all_touched=True,
        dtype=np.uint8
    )
    return mask

def display_mask(mask):
    plt.imshow(mask, cmap='gray')
    plt.axis('off')
    plt.show()


bounds = [-90.81595559689623, 14.38777463819247, -90.81324150337852, 14.39086318334812]
mask = create_mask(geojson_data, 1024, 1024, bounds)
display_mask(mask)
```
